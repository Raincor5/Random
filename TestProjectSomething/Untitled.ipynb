```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ith = 1000
jth = 5
a = np.ndarray(shape=(ith, jth))
for i in range(ith):
    for j in range(jth):
        a[i, j] = np.random.randint(1, high=50)
df = pd.DataFrame(data=a, columns=list("ABCDE"))


x = df.index.to_numpy()
A_y = df["A"]
B_y = df["B"]
plt.plot(x, A_y)
plt.plot(x, B_y, "-.")
plt.show()
```
